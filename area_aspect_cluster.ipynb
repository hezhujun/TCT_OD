```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
area_aspect = np.loadtxt("dataset/area_aspect_train.txt")
print(area_aspect.shape)
plt.scatter(area_aspect[:, 0], area_aspect[:, 1])
plt.show()
```

```python
mu = np.mean(area_aspect, axis=0)
sigma = np.std(area_aspect, axis=0)
print("mean {} std {}".format(mu, sigma))

data_norm = (area_aspect - mu) / sigma
```

```python
def cluster(data, k):
    model = KMeans(n_clusters=k)
    model = model.fit(data)
    return model, model.inertia_
```

```python
ks = [i + 1 for i in range(10)]
losses, models = [], []
for i in ks:
    model, loss = cluster(data_norm[:, 0].reshape(-1, 1), i)
    models.append(model)
    losses.append(loss)

plt.scatter(ks, losses)
plt.show()
```

```python
area_idx = 2
n_cluster = ks[area_idx]
best_model = models[area_idx]
area_centers = best_model.cluster_centers_
print(best_model.cluster_centers_ * sigma[0] + mu[0])
plt.scatter(area_aspect[:, 0], [0 for _ in range(area_aspect.shape[0])], c=best_model.labels_)
plt.show()
```

```python
area_indexs1 = best_model.labels_ == 0
data_norm1 = data_norm[area_indexs1]
losses, models = [], []
for i in ks:
    model, loss = cluster(data_norm1[:, 1].reshape(-1, 1), i)
    models.append(model)
    losses.append(loss)

plt.scatter(ks, losses)
plt.show()
```

```python
aspect_idx1 = 2
n_cluster1 = ks[aspect_idx1]
aspect_best_model1 = models[aspect_idx1]
print(aspect_best_model1.cluster_centers_ * sigma[1] + mu[1])
plt.scatter(area_aspect[area_indexs1][:, 1], [0 for _ in range(len(aspect_best_model1.labels_))], c=aspect_best_model1.labels_)
plt.show()
```
